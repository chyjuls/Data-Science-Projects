```python
# create functions.....


# using classes to load files one after the other so as to merge them successfully...

class MyCSVLoader:
    def __init__(self):
        self.airports = None
        self.airport_frequency = None
        self.runways = None

    def load_CSV(self):
        loadingCSV = fd.askopenfilename()
        # try/exception in case the wrong file format is choosen...
        try:
            return pd.read_json(loadingCSV, orient='split')
        except (ValueError, AttributeError):
            tkinter.messagebox.askretrycancel('retry', 'Failed! want to try again?')

    def load_all(self):
        try:
            self.airports = self.load_CSV()
            self.airport_frequency = self.load_CSV()
            self.runways = self.load_CSV()
        except (ValueError, AttributeError, NameError):
            tkinter.messagebox.showerror('retry', 'Quit application and start again')


csvLoader = MyCSVLoader()


# Function to analyse and display manipulated data(showing airports and frequencies data)
def clean_merge():
    global df2, dfs_rows
    global df
    global df1
    global df4
    global dfs
    global count
    try:
        df = csvLoader.airports
        df = df[['ident', 'type', 'name', 'iso_country']]
        df_mask = df['iso_country'] == 'GB'
        df = df[df_mask]
        values = ['closed', 'heliport', 'balloonport', 'seaplaneport']
        df = df[df.type.isin(values) == False]
        df = df.sort_values(by='type')

        df1 = csvLoader.airport_frequency
        df1.rename(columns={'airport_ident': 'ident'}, inplace=True)
        df1 = df1[['ident', 'frequency_mhz']]

        df4 = csvLoader.runways
        df4 = df4.dropna()
        df4 = df4[['id', 'airport_ref', 'airport_ident', 'closed']]
        # rename column to ident
        df4.rename(columns={'airport_ident': 'ident'}, inplace=True)
        # then perform a merge of all three files
        df3 = pd.merge(df, df1, on='ident')
        df3 = pd.merge(df3, df4, on='ident')
        # fill nan values with 0
        df3 = df3.fillna(0)
        # drop unneccessary columns such as closed from data...
        df3.drop(df3.columns[5:], axis=1, inplace=True)
        # extract small airport from data
        small_df = df3[df3['type'] == 'small_airport']
        # rename columns with category=='small_airports
        small_df = small_df.rename(
            columns={'type': 'Small_airport', 'frequency_mhz': 'Small_freq', 'name': 'Small_name'}, )
        small_df = small_df.drop(['iso_country'], axis=1)
        # find shape of data before merge
        # do same for medium....
        medium_df = df3[df3['type'] == 'medium_airport']
        medium_df = medium_df.rename(
            columns={'type': 'Medium_airport', 'frequency_mhz': 'Medium_freq', 'name': 'Medium_name'})
        medium_df = medium_df.drop(['iso_country'], axis=1)
        # do same for large
        large_df = df3[df3['type'] == 'large_airport']
        large_df = large_df.rename(
            columns={'type': 'Large_airport', 'frequency_mhz': 'Large_freq', 'name': 'Large_name'})
        large_df = large_df.drop(['iso_country'], axis=1)

        # perform another merge with new columns merge on ident...
        dfx = pd.merge(small_df, medium_df, on='ident', how='outer')
        # perform second merge....
        dfs = pd.merge(dfx, large_df, on='ident', how='outer')
        # fill nan with 0..
        dfs = dfs.fillna(0)

        clear_analysis_data()

        my_tree['columns'] = list(dfs.columns)
        my_tree['show'] = 'headings'

        for column in my_tree['columns']:
            my_tree.heading(column, text=column)
            dfs_rows = dfs.to_numpy().tolist()
        count = 0
        for row in dfs_rows:
            if count % 2 == 0:
                my_tree.insert('', 'end', values=row, tags=('evenrow',))
            else:
                my_tree.insert('', 'end', values=row, tags=('oddrow',))
            count += 1
        return None

    except (ValueError, IndexError, KeyError, TypeError, NameError, AttributeError) as e:
        tkinter.messagebox.showerror('warning',
                                     'An error has occurred, please choose the right file..and try again or exit the program')
        return None


def clear_analysis_data():
    my_tree.delete(*my_tree.get_children())


def file_dialog():
    global all_files

    try:

        file_name = fd.askopenfilename(initialdir=path,
                                         title='Select a file',
                                         filetypes=(
                                             ('csv files', '*.csv'), ('json files', '*json'), ('All files', '*.*')))

    except FileNotFoundError:

        tkinter.messagebox.showerror('Error', "File not found")
        tkinter.messagebox.askretrycancel('retry', 'Failed! want to try again?')
        return None

    label_file['text'] = file_name
    label_file['text'] = file_name
    label_file['text'] = file_name


def load_csv_data():
    global dfa_rows
    global dfa
    global count
    file_path = label_file['text']
    try:
        csv_filename = file_path
        dfa = pd.read_csv(csv_filename, delimiter=',')
        clear_data()

        my_tree['columns'] = list(dfa.columns)
        my_tree['show'] = 'headings'
        for column in my_tree['columns']:
            my_tree.heading(column, text=column)
            dfa_rows = dfa.to_numpy().tolist()
        count = 0
        for row in dfa_rows:
            if count % 2 == 0:
                my_tree.insert('', 'end', values=row, tags=('evenrow',))
            else:
                my_tree.insert('', 'end', values=row, tags=('oddrow',))
            count += 1
        return None

    except ValueError:
        tkinter.messagebox.showerror('Error', 'File is invalid')
        return None
    except FileNotFoundError:
        tkinter.messagebox.askretrycancel('retry', 'Failed! want to try again?')


def clear_data():
    my_tree.delete(*my_tree.get_children())


def clean_runway():
    global df4, df4_rows
    file_path = label_file['text']
    try:
        csv_filename = file_path
        pd.set_option('display.max_rows', None, 'display.max_columns', 7, 'max_colwidth', 20)
        df4 = pd.read_csv(csv_filename, delimiter=',')
        df4 = df4[['id', 'airport_ref', 'airport_ident', 'closed']]
        df4.rename(columns={'airport_ident': 'ident'}, inplace=True)
        df4 = df4.dropna()

        clear_runway_data()

        my_tree['columns'] = list(df4.columns)
        my_tree['show'] = 'headings'
        for column in my_tree['columns']:
            my_tree.heading(column, text=column)

            df4_rows = df4.to_numpy().tolist()

        count = 0
        for row in df4_rows:
            if count % 2 == 0:
                my_tree.insert('', 'end', values=row, tags=('evenrow',))
            else:
                my_tree.insert('', 'end', values=row, tags=('oddrow',))
            count += 1
        return None

    except FileNotFoundError:
        tkinter.messagebox.showerror('Error', "File not found")
        return None
    except (ValueError, KeyError):
        tkinter.messagebox.showerror('Error', "You have chosen the wrong file, please  try again")
        tkinter.messagebox.showwarning('Warning', "DO NOT SAVE FILE, CANCEL AND CLOSE ALL INTERACTIONS")
    return None


def clear_runway_data():
    my_tree.delete(*my_tree.get_children())


def clean_frequency_data():
    file_path = label_file['text']
    try:
        csv_filename = file_path
        pd.set_option('display.max_rows', None, 'display.max_columns', None, 'max_colwidth', 10)
        df1 = pd.read_csv(csv_filename, delimiter=',')
        df1 = df1.dropna()
        df1.rename(columns={'airport_ident': 'ident'}, inplace=True)
        df1 = df1[['ident', 'frequency_mhz']]

        clear_frequency_data()

        my_tree['columns'] = list(df1.columns)
        my_tree['show'] = 'headings'
        for column in my_tree['columns']:
            my_tree.heading(column, text=column)

            df1_rows = df1.to_numpy().tolist()

        count = 0
        for row in df1_rows:
            if count % 2 == 0:
                my_tree.insert('', 'end', values=row, tags=('evenrow',))
            else:
                my_tree.insert('', 'end', values=row, tags=('oddrow',))
            count += 1
        return None


    except FileNotFoundError:
        tkinter.messagebox.showerror('Error', "File not found")
        return None
    except (ValueError, KeyError):
        tkinter.messagebox.showerror('Error', "You have chosen the wrong file, please  try again")
        tkinter.messagebox.showwarning('Warning', "DO NOT SAVE FILE, CANCEL AND CLOSE ALL INTERACTIONS")
        return None


def clear_frequency_data():
    my_tree.delete(*my_tree.get_children())


def clean_airport_data():
    file_path = label_file['text']
    try:
        csv_filename = file_path
        pd.set_option('display.max_rows', None, 'display.max_columns', None, 'max_colwidth', 20)
        df = pd.read_csv(csv_filename, delimiter=',', usecols=['ident', 'type', 'name', 'iso_country'])
        df = df.dropna()
        df_mask = df['iso_country'] == 'GB'
        df = df[df_mask]
        values = ['closed', 'heliport', 'balloonport']
        df = df[df.type.isin(values) == False]
        df = df.sort_values(by='type')
        df = df.fillna(0)

        clear_airport_data()

        my_tree['columns'] = list(df.columns)
        my_tree['show'] = 'headings'
        for column in my_tree['columns']:
            my_tree.heading(column, text=column)
            df_rows = df.to_numpy().tolist()

        count = 0
        for row in df_rows:
            if count % 2 == 0:
                my_tree.insert('', 'end', values=row, tags=('evenrow',))
            else:
                my_tree.insert('', 'end', values=row, tags=('oddrow',))
            count += 1
        return None

    except FileNotFoundError:
        tkinter.messagebox.showerror('Error', "File not found")
        return None
    except (ValueError, KeyError):
        tkinter.messagebox.showerror('Error', "You have chosen the wrong file, please  try again")
        tkinter.messagebox.showwarning('Warning', "DO NOT SAVE FILE, CANCEL AND CLOSE ALL INTERACTIONS")
        return None


def clear_airport_data():
    my_tree.delete(*my_tree.get_children())


path = './'


def write_JSONFile(path, fileName, data):
    data.to_json(path, orient='split', index=False, indent=4)


def onClick():
    tkinter.messagebox.showinfo('Please load the JSON file for analysis')


def save_tojson():
    try:
        global df, files, fileName
        global data
        data = {}
        data = dfa
        if messagebox.askyesno('Confirmation', 'Are you sure you want to save data as json file?'):
            files = [('JSON File', '*.json')]
            fileName = 'Myfile'
        file_position = asksaveasfile(filetypes=files, defaultextension=json, initialfile='Myfile')
        write_JSONFile(file_position, fileName, data)
    except (ValueError, NameError):
        tkinter.messagebox.showerror('Error', 'Please select a csv data file, clean data and then save as a JSON file')
    except FileNotFoundError:
        tkinter.messagebox.showerror('Error', 'File not found')


def show_small():
    global  small_data,  small_100_rows, small_100
    try:
        # filter out aitports to display data susbsets by creating a subset and filtering out the zeros
        small_data = dfs[['ident', 'Small_airport', 'Small_freq']]
        small_data = small_data[(small_data.Small_freq != 0)]
        # using pd series to find the mode
        mode = lambda x: pd.Series.mean(pd.Series.mode(x))
        # applying slicing to get medium frequencies greater than 100
        small_100 = small_data.loc[small_data['Small_freq'] > 100].agg(['mean', 'median', mode])
        small_100 = small_100.round(decimals=2)
        # transforming the data frame(swapping rows and columns
        small_100 = small_100.T
        small_100 = small_100.rename(columns={'<lambda>': 'mode'})


    except (ValueError, FileNotFoundError, NameError):
        tkinter.messagebox.showerror('Error',
                                     "You have not selected any data files for analysis, "
                                     "please select a file by pressing the load data for analysis button.")
    clear_small()
    try:
        my_tree['columns'] = list(small_100.columns)
        my_tree['show'] = 'headings'
        for column in my_tree['columns']:
            my_tree.heading(column, text=column)

            small_100_rows = small_100.to_numpy().tolist()

        count = 0
        for row in small_100_rows:
            if count % 2 == 0:
                my_tree.insert('', 'end', values=row, tags=('evenrow',))
            else:
                my_tree.insert('', 'end', values=row, tags=('oddrow',))
            count += 1
        return None
    except (ValueError, FileNotFoundError, NameError):
        tkinter.messagebox.showerror('Error',
                                     "You have not selected any data files for analysis, "
                                     "please select a file by pressing the load data for analysis button.")
    print(small_100)


def clear_small():
    my_tree.delete(*my_tree.get_children())


def show_medium():
    global medium_data, medium_100_rows, medium_100
    try:
        # filter out airports to display data subsets by creating a subset and filtering out the zeros
        medium_data = dfs[['ident', 'Medium_airport', 'Medium_freq']]
        medium_data = medium_data[(medium_data.Medium_freq != 0)]
        # using pd series to find the mode
        mode = lambda x: pd.Series.mean(pd.Series.mode(x))
        # applying slicing to get medium frequencies greater than 100
        medium_100 = medium_data.loc[medium_data['Medium_freq'] > 100].agg(['mean', 'median', mode])
        medium_100 = medium_100.round(decimals=2)
        # transforming the data frame(swapping rows and columns
        medium_100 = medium_100.T
        medium_100 = medium_100.rename(columns={'<lambda>': 'mode'})
    except (ValueError, FileNotFoundError, NameError):
        tkinter.messagebox.showerror('Error',
                                     "You have not selected any data files for analysis, "
                                     "please select a file by pressing the load data for analysis button.")
    
    clear_medium()

    my_tree['columns'] = list(medium_100.columns)
    my_tree['show'] = 'headings'
    for column in my_tree['columns']:
        my_tree.heading(column, text=column)

        medium_100_rows = medium_100.to_numpy().tolist()

    count = 0
    for row in medium_100_rows:
        if count % 2 == 0:
            my_tree.insert('', 'end', values=row, tags=('evenrow',))
        else:
            my_tree.insert('', 'end', values=row, tags=('oddrow',))
        count += 1
    return None


def clear_medium():
    my_tree.delete(*my_tree.get_children())


def show_all():

    global df3_rows, df3
    try:
        df = csvLoader.airports
        df = df[['ident', 'type', 'name', 'iso_country']]
        df_mask = df['iso_country'] == 'GB'
        df = df[df_mask]
        values = ['closed', 'heliport', 'balloonport', 'seaplaneport']
        df = df[df.type.isin(values) == False]
        df = df.sort_values(by='type')

        df1 = csvLoader.airport_frequency
        df1.rename(columns={'airport_ident': 'ident'}, inplace=True)
        df1 = df1[['ident', 'frequency_mhz']]

        df4 = csvLoader.runways
        df4 = df4.dropna()
        df4 = df4[['id', 'airport_ref', 'airport_ident', 'closed']]
        # rename column to ident
        df4.rename(columns={'airport_ident': 'ident'}, inplace=True)
        # then perform a merge of all three files
        df3 = pd.merge(df, df1, on='ident')
        df3 = pd.merge(df3, df4, on='ident')
        # fill nan values with 0
        df3 = df3.fillna(0)
        # drop unneccessary columns such as closed from data...
        df3.drop(df3.columns[5:], axis=1, inplace=True)

    except (ValueError, FileNotFoundError, NameError):
        tkinter.messagebox.showerror('Error',
                                     "You have not selected any data files for analysis, "
                                     "please select a file by pressing the load data for analysis button.")

    clear_large()

    my_tree['columns'] = list(df3.columns)
    my_tree['show'] = 'headings'
    for column in my_tree['columns']:
        my_tree.heading(column, text=column)

        df3_rows = df3.to_numpy().tolist()

    count = 0
    for row in df3_rows:
        if count % 2 == 0:
            my_tree.insert('', 'end', values=row, tags=('evenrow',))
        else:
            my_tree.insert('', 'end', values=row, tags=('oddrow',))
        count += 1
    return None


def clear_large():
    my_tree.delete(*my_tree.get_children())


def large_frequency():
    global dfs, large_rows, large

    try:
        # filter out aitports to display data susbsets by creating a subset and filtering out the zeros
        large_data = dfs[['ident', 'Large_airport', 'Large_freq']]
        large_data = large_data[(large_data.Large_freq != 0)]
        # getting the mean of two values that will be calculated for the mode, when you use a pd series....
        mode = lambda x: pd.Series.mean(pd.Series.mode(x))

        large = large_data.groupby('Large_airport').Large_freq.agg(['mean', 'median', mode])
        large = large.rename(columns={'<lambda_0>': 'mode'})
        # round to 2 decimal places
        large = large.round(decimals=2)


    except (ValueError, FileNotFoundError, NameError):
        tkinter.messagebox.showerror('Error',
                                     "You have not selected any data files for analysis, "
                                     "please select a file by pressing the load data for analysis button.")

    
    clear_stats_data()

    my_tree['columns'] = list(large.columns)
    my_tree['show'] = 'headings'
    for column in my_tree['columns']:
        my_tree.heading(column, text=column)
        large_rows = large.to_numpy().tolist()

    count = 0
    for row in large_rows:
        if count % 2 == 0:
            my_tree.insert('', 'end', values=row, tags=('evenrow',))
        else:
            my_tree.insert('', 'end', values=row, tags=('oddrow',))
        count += 1

    return None


def clear_stats_data():
    my_tree.delete(*my_tree.get_children())


def show_large100():
    global large_100_rows, large_100
    try:
        # all large airport frequencies greater than 100
        large_data = dfs[['ident', 'Large_airport', 'Large_freq']]
        large_data = large_data[(large_data.Large_freq != 0)]
        # get the mode using lambda and pd.series...
        mode = lambda x: pd.Series.mean(pd.Series.mode(x))
        large_100 = large_data.loc[large_data['Large_freq'] > 100].agg(['mean', 'median', mode])
        large_100 = large_100.round(decimals=2)
        large_100 = large_100.T
        large_100 = large_100.rename(columns={'<lambda>': 'mode'})
    except (ValueError, FileNotFoundError, NameError):
        tkinter.messagebox.showerror('Error',
                                     "You have not selected any data files for analysis, "
                                     "please select a file by pressing the load data for analysis button.")

    clear_large100()

    my_tree['columns'] = list(large_100.columns)
    my_tree['show'] = 'headings'
    for column in my_tree['columns']:
        my_tree.heading(column, text=column)

        large_100_rows = large_100.to_numpy().tolist()

    count = 0
    for row in large_100_rows:
        if count % 2 == 0:
            my_tree.insert('', 'end', values=row, tags=('evenrow',))
        else:
            my_tree.insert('', 'end', values=row, tags=('oddrow',))
        count += 1
    return None


def clear_large100():
    my_tree.delete(*my_tree.get_children())


def plot_graph():
    try:

        small_data = dfs[['ident', 'Small_airport', 'Small_freq']]
        small_data = small_data[(small_data.Small_freq != 0)]
        small_freq = small_data['Small_freq']
        # values in a numpy array
        small_freq = np.array(small_freq)
        # get small_airport counts
        # put in an index of 1 to 22
        small_data.index = np.arange(1, len(small_data) + 1)
        # get small airport frequencies from cleaned and grouped data: 'Small data'

        colors = sns.color_palette("pastel")
        coord1 = 121
        coord2 = 122
        fig = plt.figure('Small Airport Frequency', figsize=(8, 4))
        plt.subplot(coord1)
        sns.set_style("darkgrid")
        x = small_data.index
        y = small_freq
        plt.xlabel('Small airport', fontweight='bold', size=9)
        plt.ylabel('Frequency_mhz', fontweight='bold', size=9)
        plt.title("Fig 1.0 Bar Chart", fontweight='bold')
        plt.bar(x, y, color=colors[:5], width=1)
        l1 = np.arange(0, 25, 5)
        l2 = np.arange(0, 300, 20)
        plt.xticks(l1)
        plt.yticks(l2)
        plt.xlim(1, 25)
        plt.ylim(1, 300)

        plt.subplot(coord2)
        sns.set_style("darkgrid")
        small_data = dfs[['Small_airport', 'Small_freq']]
        small_data = small_data[(small_data.Small_freq != 0)]
        small_data.index = np.arange(1, len(small_data) + 1)
        new_data = {'small_airport': small_data.index, 'Frequencies': small_freq}
        new_data = pd.DataFrame(new_data)
        plt.title("Fig 1.1 Scatter Plot", fontweight='bold')
        plt.xlabel("small_airport", fontweight='bold', size=9)
        plt.ylabel("Frequencies", fontweight='bold', size=9)
        sns.scatterplot(x='small_airport', y='Frequencies', data=new_data, palette=colors[:5], edgecolor="black")
        plt.tight_layout()
        plt.show()

    except (ValueError, FileNotFoundError, NameError):
        tkinter.messagebox.showerror('Error',
                                     "You have not selected any data files for analysis, "
                                     "please select a file by pressing the load data for analysis button.")


def correlation():
    try:
        ##plot heatmap to show correlation
        # plt.figure(figsize=(10, 2))
        fig = plt.figure('Frequency Communications Correlation',figsize=(6, 4))
        corr = dfs.corr()
        corr = corr.round(decimals=2)
        # sns.heatmap(corr_1,annot=True)
        # shows how closely related the frequencies are for the airports, the colours going from blue to red...
        sns.heatmap(corr, annot=True, vmax=1, vmin=-1, center=0, cmap=sns.diverging_palette(20, 220, n=200))
        plt.show()
    except (NameError, ValueError, FileNotFoundError):
        tkinter.messagebox.showerror('Error',
                                     "You have not selected any data files for analysis, "
                                     "please select a file by pressing the load data for analysis button.")


def corr_table():
    global corr_rows
    global corr
    try:
    #....now to find correlation...
    #reset index ...to enable proper data matrix
    # correlation is 1, when a value correlates to itself...

        corr = dfs.corr()
        corr = corr.round(decimals=2)
        corr = corr.reset_index()
        clear_corr()

        my_tree['columns'] = list(corr.columns)
        my_tree['show'] = 'headings'
        for column in my_tree['columns']:
            my_tree.heading(column, text=column)

            corr_rows = corr.to_numpy().tolist()

        count = 0
        for row in corr_rows:
            if count % 2 == 0:
                my_tree.insert('', 'end', values=row, tags=('evenrow',))
            else:
                my_tree.insert('', 'end', values=row, tags=('oddrow',))
            count += 1
        return None



    except (NameError, ValueError, FileNotFoundError):
        tkinter.messagebox.showerror('Error',
                                     "You have not selected any data files for analysis, "
                                     "please select a file by pressing the load data for analysis button.")


def clear_corr():
    my_tree.delete(*my_tree.get_children())


# function to clear tree view of previous data
def clear_table():
    my_tree.delete(*my_tree.get_children())
```
